# Breaking Down The Neural Network!


### 1. Data
  The type of data a neural network processes varies drastically based on the problem being solved. When we build a neural network, we define what shape and kind of data it can accept. It may sometimes be neccessary to modify our dataset so that it can be passed to our neural network.

### 2. Layers
  As we mentioned earlier each neural network consists of multiple layers. At each layer a different transformation of data occurs. Our initial input data is fed through the layers and eventually arrives at the output layer where we will obtain the result.

  * Input Layer:
  The input layer is the layer that our initial data is passed to. It is the first layer in our neural network.

  * Output Layer:
  The output layer is the layer that we will retrive our results from. Once the data has passed through all other layers it will arrive here.

  * Hidden Layer(s):
  All the other layers in our neural network are called "hidden layers". This is because they are hidden to us, we cannot observe them. Most neural networks consist of at least one hidden layer but can have an unlimited amount. Typically, the more complex the model the more hidden layers.

![](https://miro.medium.com/v2/resize:fit:640/format:webp/0*BSxP3AHxBe_IevHC.png)

### 3. Neurons

Each layer is made up of what are called neurons. For example, say we want to pass an image that is 28x28 pixels, thats 784 pixels. We would need 784 neurons in our input layer to capture each of these pixels.

### 4. Weights
  Weights are associated with each connection in our neural network. Every pair of connected nodes will have one weight that denotes the strength of the connection between them. The model will try to determine what these weights should be to achieve the best result. Weights start out at a constant or random value and will change as the network sees training data.

### 5. Biases

  A bias is simply a constant value associated with each layer. It can be thought of as an extra neuron that has no connections. The purpose of a bias is to shift an entire activation function by a constant value. This allows a lot more flexibllity when it comes to choosing an activation and training the network. There is one bias for each layer.

![](https://miro.medium.com/v2/resize:fit:960/1*0lejoYyyQWjYzEP_BNW2nw.jpeg)

### 6. Activation Function

Activation functions are simply a function that is applied to the weighed sum of a neuron. They can be anything we want but are typically higher order/degree functions that aim to add a higher dimension to our data. We would want to do this to introduce more complexity to our model.


A list of some common activation functions and their graphs can be seen below:

* #### Sigmoid function:

  Transform $ (- \infty $ to $ \infty) $ into (0 to 1) range
  $$
  sigmoid(x) = \frac{1}{1 + e^{-x}}
  $$

* #### Tanh function:

  Similar to sigmoid, difference being that output is -1 to +1


$$
tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

  

* #### ReLU

  $$
  f(x) = \begin{cases} %
                      0 & if \; x<0 \\
                      x &  if \;x \geq 0.
                  \end{cases}
  $$

## How it Works

A neural network consists of many Nodes (Neurons) in many layers. Each layer can have any number of nodes and a neural network can have any number of layers.



Lets take the example of whats going on with a single node in the network.

![](https://miro.medium.com/v2/resize:fit:720/format:webp/1*SaQMHTLi4C7MIA4IzjAXJw.png)






\begin{equation}
Y =(\sum_{i=0}^n w_i x_i) + b
\end{equation}

 * w: stands for the weight of each connection to the neuron

 * x: stands for the value of the connected neuron from the previous value

 * b: stands for the bias at each layer, this is a constant

 * n: is the number of connections

 * Y: is the output of the current neuron


 The equation you just read is called a weighed sum. We will take this weighted sum at each and every neuron as we pass information through the network. Then we will add what's called a bias to this sum. The bias allows us to shift the network up or down by a constant value. It is like the y-intercept of a line.


 But that equation is the not complete one! We forgot a crucial part, the **activation function**. Our new equation with the addition of an activation function
 is seen below.

\begin{equation}
 Y =F((\sum_{i=0}^n w_i x_i) + b)
\end{equation}

# Forward Propogation.

* Forward propagation is the process in which input data is processed through the neural network's layers to produce an output.

* It involves passing the input data through each layer of the neural network, applying the layer's weights and activation functions, until the final output is obtained.

# Backpropagation
Backpropagation is the fundemental algorithm behind training neural networks. It is what changes the weights and biases of our network. To fully explain this process, we need to learn something called a cost/loss function.

# Loss/Cost Function

For our training data we have the features (input) and the labels (expected output), because of this we can compare the output from our network to the expected output. Based on the difference between these values we can determine if our network has done a good job or poor job. If the network has done a good job, we'll make minor changes to the weights and biases. If it has done a poor job our changes may be more drastic.

Some common loss/cost functions include.

* Mean Squared Error
\begin{equation}
y = \sum_{i=1}^{D}(x_i-y_i)^2
\end{equation}


* Mean Absolute Error
\begin{equation}y = \sum_{i=1}^{D}|x_i-y_i|
\end{equation}
* Hinge Loss
\begin{equation}
y = max(0, 1 - y \cdot \hat{y})
\end{equation}

Where 'D' represents the number of samples in the dataset.

# Optimizer
Optimization function is simply the function that implements the backpropagation algorithm described above. Here's a list of a few common ones.

* Gradient Descent
* Stochastic Gradient Descent
* Mini-Batch Gradient Descent
* Momentum
* Nesterov Accelerated Gradient

![](https://i.pinimg.com/originals/6f/d6/22/6fd62253592b42795c48dc570a17579c.gif)

During backpropagation we calculate the total error at the output nodes and propagate these errors back through the network using Backpropagation to calculate the gradients. Then we use an optimization method such as Gradient Descent to adjust all weights in the network with an aim of reducing the error at the output layer.



## Building a Nueral Network

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import random
```

```python
# load the dataset

iris = load_iris()
data, labels = iris.data[:,0:2], iris.data[:,2]

num_samples = len(labels)  # size of our dataset

# shuffle the dataset
shuffle_order = np.random.permutation(num_samples)
data = data[shuffle_order, :]
labels = labels[shuffle_order]
```

Like the 1-dimensional problem previously, we can still do linear regression, except now we have two variables and therefore two weights as well. Let's denote the input variables as x1 and x2 and instead of using m as the coefficient variable, let's use w1 and w2. So for linear regression, we would have the following function:

$$
f(X) = w_1 x_1 + w_2 x_2 + b
$$


For example, suppose set w = [0.2, 0.6] and b = -0.3. Let's calculate the resulting . We can program this as a function called "weighted_sum".

```python
def weighted_sum(x, w, b):
    return b + np.dot(w, x)

# set our paramters - weights and bias
w = [0.2, 0.6]
b = -0.3

# for example, let's use the first data point
X, y = data, labels

pred_y = [weighted_sum(x, w, b) for x in X]

# let's print out the first prediction
print("for x=[%0.2f, %0.2f], predicted = %0.2f, actual = %0.2f" % (X[0][0], X[0][1], pred_y[0], y[0]))
```

We can evaluate the quality of our predictions using cost/loss functions. Lets use the sum-squared error function



```python
# sum squared error
def cost_function(y_pred, y_actual):
    return 0.5 * np.sum((y_actual-y_pred)**2)

error = cost_function(pred_y, y)
print(error)
```

```python
# normalizing the data
X = X / np.amax(X, axis=0)
y = y / np.amax(y, axis=0)

# randomly initializing w, b
w, b = [random.random(), random.random()], random.random()

# our function w*x + b
def F(X, w, b):
    return np.sum(w*X, axis=1) + b

# calculating error using cost function(Here we use Mean Squared Error)
y_pred = F(X, w, b)
init_cost = cost_function(y_pred, y)

print("initial parameters: w1=%0.3f, w2=%0.3f, b=%0.3f"%(w[0], w[1], b))
print("initial cost = %0.3f" % init_cost)
```

Now the partial derivatives are:

$$
\frac{\partial J}{\partial w_1} = - \sum{x_1^i \cdot (y^i - (w_1 x_1^i + w_2 x_2^i+ b))}
$$

$$
\frac{\partial J}{\partial w_2} = - \sum{x_2^i \cdot (y^i - (w_1 x_1^i + w_2 x_2^i+ b))}
$$

$$
\frac{\partial J}{\partial b} = - \sum{y^i - (w_1 x_1^i + w_2 x_2^i+ b)}
$$

```python
# implement partial derivatives of our parameters

def dJdw1(X, y, w, b):
    return -np.dot(X[:,0], y - F(X, w, b))

def dJdw2(X, y, w, b):
    return -np.dot(X[:,1], y - F(X, w, b))

def dJdb(X, y, w, b):
    return -np.sum(y - F(X, w, b))
```

Aftet that, we use the following update rule, where we calculate the gradient and then adjust the parameters.

$$
w_1 = w_1 - \alpha \cdot \frac{\partial J}{\partial w_i}
$$

$$
w_2 = w_2 - \alpha \cdot \frac{\partial J}{\partial w_2}
$$

$$
b = b - \alpha \cdot \frac{\partial J}{\partial b}
$$

```python
# choose the learning rate parameter and number of iterations
lr = 0.001
n_iters = 2000

# run through gradient descent
errors = []
for i in range(n_iters):
    w[0] = w[0] - lr * dJdw1(X, y, w, b)
    w[1] = w[1] - lr * dJdw2(X, y, w, b)
    b = b - lr * dJdb(X, y, w, b)
    y_pred = F(X, w, b)
    j = cost_function(y_pred, y)
    errors.append(j)
```

```python
# plot the error
plt.plot(range(n_iters), errors, linewidth=2)
plt.title("Cost by iteration")
plt.ylabel("Cost")
plt.xlabel("iterations")
```

* At the beginning of training, the loss is typically high as the model's weights are randomly initialized,the model struggles to make accurate predictions, resulting in a high training loss.

* As the training progresses, the loss generally decreases, the model adjusts its weights and biases to minimize the difference between its predictions and the actual target values.

Let's implement the sigmoid function.



```python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))
```

Basically, a sigmoid just squashes any input it gets to between 0 and 1. So we now modify our basic function in the following way: instead of just outputting the weighted sum by itself,we now pass that through the sigmoid function.

  So instead of y = (w_1 * x_1) + (w_2 * x_2) + b our function now looks like:

$$
y = \frac{1}{1 + e^{-(w_1 x_1 + w_2 x_2 + b)}}
$$

```python
def weighted_sum(x, w, b):
    return b + np.dot(w, x)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# our parameters
w = [0.2, 0.6]
b = -0.3

X, y = data, labels

# get weighted sum like before
Z = [weighted_sum(x, w, b) for x in X]

# now transform the weighted sums with a sigmoid
y_pred = [sigmoid(z) for z in Z]

# evaluate error
error = cost_function(y_pred, y)
print(error)
```

The function we posed above can already be considered a neural network. But let's complicate things a bit further, by adding a hidden layer. Neurons can be arranged in layers. So instead of having just two input neurons and an output neuron, let's place a layer of three neurons in the middle

![](https://raw.githubusercontent.com/ml4a/ml4a/a8831f15b581f091d16003b0b61a68ed1bbbb770/assets/neuralnet.jpg)

```python
W1 = np.random.randn(2, 3)
W2 = np.random.randn(3, 1)

print("W1=", W1)
print("W2=", W2)
```

#### implementing a forward pass

```python

X, y = data, labels

# first layer weighted sum z
z = np.dot(X, W1)

# project z through non-linear sigmoid
z = sigmoid(z)

# do another dot product at end (sigmoid is omitted)
y_pred = np.dot(z, W2)

# what is our cost
error = cost_function(y_pred, y)

print('predicted %0.2f for example 0, actual %0.2f, total cost %0.2f'%(pred_y[0], y[0], error))
```

Let's turn the above operations into a class.



```python
class Neural_Network(object):
    def __init__(self, n0, n1, n2):
        self.n0 = n0
        self.n1 = n1
        self.n2 = n2

        # initialize weights
        self.W1 = np.random.randn(self.n0, self.n1)
        self.W2 = np.random.randn(self.n1 ,self.n2)

    def predict(self, x):
        z = np.dot(x, self.W1)
        z = sigmoid(z)
        y = np.dot(z, self.W2)
        return y
```

```python
# instantiate a neural network with 2 input neurons, 3 hidden neurons, and 1 output neuron using above class

net = Neural_Network(2, 3, 1)
```

Now to do a forward pass, we can simply run the networks predict function:


```python
X, y = data, labels
y_pred = net.predict(X)
error = cost_function(y_pred, y)

print('predicted %0.2f for example 0, actual %0.2f, total cost %0.2f'%(pred_y[0], y[0], error))
```

 we have a 2x3x1 neural network with 9 weights and 4 biases for 13 total parameters.

 Now we optimize the parameters to minimize our cost function using Gradient Descent.

Gradient Descent will find the gradient of the cost/loss function(J)
 with respect to the parameters w,b.

 $$
 w_i := w_i - \alpha \cdot \frac{\partial J}{\partial w_i}
 $$

```python
def get_gradient(net, X, y):
    w_delta = 1e-8

    # get the current value of the loss, wherever the parameters are
    y_pred_current = net.predict(X)
    error_current = cost_function(y_pred_current, y)

    # grab the current weights and copy them (so we can restore them after modification)
    dw1, dw2 = np.zeros((net.n0, net.n1)), np.zeros((net.n1, net.n2))
    W1, W2 = np.copy(net.W1), np.copy(net.W2)

    # Calculate gradient for the first layer
    for i in range(net.n0):
        for j in range(net.n1):
            net.W1 = np.copy(W1)
            net.W1[i][j] += w_delta
            y_pred = net.predict(X)
            error = cost_function(y_pred, y)
            dw1[i][j] = (error - error_current) / w_delta

    # Calculate gradient for the second layer
    for i in range(net.n1):
        for j in range(net.n2):
            net.W2 = np.copy(W2)
            net.W2[i][j] += w_delta
            y_pred = net.predict(X)
            error = cost_function(y_pred, y)
            dw2[i][j] = (error - error_current) / w_delta

    # restore the original weights
    net.W1, net.W2 = np.copy(W1), np.copy(W2)

    return dw1, dw2
```

The above function 'get_gradient' calculates the gradient of a 2-layer network net, for our dataset X, y

Now we train the network by the following steps:

1. Load our dataset
2. Instantiate a neural network
3. Train it on the data using the gradient method made above.

```python
# load the data and labels
X, y = data, labels.reshape((len(labels),1))

# it's always a good idea to normalize the data between 0 and 1
X = X/np.amax(X, axis=0)
y = y/np.amax(y, axis=0)

# create a 2x3x1 neural net
net = Neural_Network(2, 3, 1)

# what is the current cost?
y_orig = net.predict(X)
init_cost = cost_function(y_orig, y)
print("initial cost = %0.3f" % init_cost)

# Set the learning rate, and how many epochs (updates) to try
n_epochs = 2000
learning_rate = 0.01

# for each epoch, calculate the gradient, then subtract it from the parameters, and save the cost
errors = []
for i in range(n_epochs):
    dw1, dw2 = get_gradient(net, X, y)
    net.W1 = net.W1 - learning_rate * dw1
    net.W2 = net.W2 - learning_rate * dw2
    y_pred = net.predict(X)
    error = cost_function(y_pred, y)
    errors.append(error)
```

```python
# plotting iterations vs error
plt.plot(range(0, len(errors)), errors)
plt.xlabel('Iterations')
plt.ylabel('Error')
```

```python
import numpy as np
from sklearn.datasets import load_iris

# Load and shuffle the dataset
iris = load_iris()
data, labels = iris.data[:,0:2], iris.data[:,2]

num_samples = len(labels)  # size of our dataset

# shuffle the dataset
shuffle_order = np.random.permutation(num_samples)
data = data[shuffle_order, :]
labels = labels[shuffle_order]

# Now data and labels are defined. Assign to X, y for consistent variable names.
X, y = data, labels.reshape((len(labels),1))

# it's always a good idea to normalize the data between 0 and 1
X = X/np.amax(X, axis=0)
y = y/np.amax(y, axis=0)

# Define the sigmoid function
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

# Define the Neural_Network class
class Neural_Network(object):
    def __init__(self, n0, n1, n2):
        self.n0 = n0
        self.n1 = n1
        self.n2 = n2

        # initialize weights
        self.W1 = np.random.randn(self.n0, self.n1)
        self.W2 = np.random.randn(self.n1 ,self.n2)

    def predict(self, x):
        z = np.dot(x, self.W1)
        z = sigmoid(z)
        y = np.dot(z, self.W2)
        return y

def cost_function(y_pred, y_actual):
    return 0.5 * np.sum((y_actual-y_pred)**2)

# create a 2x3x1 neural net
net = Neural_Network(2, 3, 1)

# what is the current cost?
y_orig = net.predict(X)
init_cost = cost_function(y_orig, y)

# Define get_gradient
def get_gradient(net, X, y):
    w_delta = 1e-8

    # get the current value of the loss, wherever the parameters are
    y_pred_current = net.predict(X)
    error_current = cost_function(y_pred_current, y)

    # grab the current weights and copy them (so we can restore them after modification)
    dw1, dw2 = np.zeros((net.n0, net.n1)), np.zeros((net.n1, net.n2))
    W1, W2 = np.copy(net.W1), np.copy(net.W2)

    # Calculate gradient for the first layer
    for i in range(net.n0):
        for j in range(net.n1):
            net.W1 = np.copy(W1)
            net.W1[i][j] += w_delta
            y_pred = net.predict(X)
            error = cost_function(y_pred, y)
            dw1[i][j] = (error - error_current) / w_delta

    # Calculate gradient for the second layer
    for i in range(net.n1):
        for j in range(net.n2):
            net.W2 = np.copy(W2)
            net.W2[i][j] += w_delta
            y_pred = net.predict(X)
            error = cost_function(y_pred, y)
            dw2[i][j] = (error - error_current) / w_delta

    # restore the original weights
    net.W1, net.W2 = np.copy(W1), np.copy(W2)

    return dw1, dw2

# Set the learning rate, and how many epochs (updates) to try
n_epochs = 2000
learning_rate = 0.01

# for each epoch, calculate the gradient, then subtract it from the parameters, and save the cost
for i in range(n_epochs):
    dw1, dw2 = get_gradient(net, X, y)
    net.W1 = net.W1 - learning_rate * dw1
    net.W2 = net.W2 - learning_rate * dw2
    y_pred = net.predict(X)

# Define mean_absolute_error (assuming it's defined elsewhere or explicitly here if needed)
def mean_absolute_error(y_pred, y_actual):
    return np.sum(np.abs(y_actual - y_pred)) / len(y_actual)

# Now y_pred and y are defined for the final calculation
mae_error = mean_absolute_error(y_pred, y)
print("Mean Absolute Error: %0.3f" % mae_error)
```

#### More on Sigmoid function:

*   **Use Cases**: Historically popular for output layers in binary classification problems where the output needs to be interpreted as a probability (between 0 and 1).
*   **Advantages**: Smooth gradient, produces a clear probability interpretation.
*   **Drawbacks**:
    *   **Vanishing Gradient Problem**: For very large positive or negative inputs, the sigmoid function saturates, meaning its derivative becomes very close to zero. This can cause the gradients to become extremely small during backpropagation, effectively stopping the learning process for layers closer to the input.
    *   **Outputs Not Zero-Centered**: The output of the sigmoid function is always positive (between 0 and 1). This can lead to a 'zig-zagging' effect in gradient descent updates for the weights, making optimization less efficient.

#### More on Tanh function:

*   **Use Cases**: Commonly used in hidden layers of neural networks, especially when the output needs to be centered around zero. It's an alternative to sigmoid that often performs better.
*   **Advantages**:
    *   **Zero-Centered Output**: Unlike sigmoid, Tanh's output ranges from -1 to 1, which helps in centering the data and often makes the training of the next layer easier and more efficient.
    *   **Stronger Gradients**: The gradients are typically steeper than sigmoid, leading to faster learning in some cases.
*   **Drawbacks**:
    *   **Vanishing Gradient Problem**: Similar to sigmoid, Tanh also suffers from the vanishing gradient problem when inputs are very large or very small, as its derivative approaches zero in these regions.

#### More on ReLU function:

*   **Use Cases**: Currently the most popular activation function for hidden layers in deep neural networks. It avoids the vanishing gradient problem for positive inputs.
*   **Advantages**:
    *   **Solves Vanishing Gradient Problem (for positive inputs)**: For positive inputs, the derivative is always 1, which prevents the gradient from vanishing. This allows for faster training of deep networks.
    *   **Computational Efficiency**: ReLU involves simple operations (max(0, x)), making it computationally very efficient compared to sigmoid and Tanh, which involve exponentials.
    *   **Sparsity**: It can lead to sparse activations, meaning some neurons output 0, which can be beneficial for certain tasks.
*   **Drawbacks**:
    *   **Dying ReLU Problem**: For negative inputs, the derivative is 0. This means that once a neuron's output becomes negative, it can stop learning completely, as its gradient will always be zero. This is known as the 'dying ReLU' problem.
    *   **Not Zero-Centered**: Like sigmoid, ReLU's output is not zero-centered, which can also lead to zig-zagging gradient updates.


#### More on Mean Squared Error (MSE) / Sum Squared Error (SSE):

*   **Formula**: $$\sum_{i=1}^{D}(x_i-y_i)^2$$ or $$\frac{1}{D}\sum_{i=1}^{D}(x_i-y_i)^2$$
*   **Use Cases**: Primarily used in **regression problems** where the goal is to predict a continuous value. It penalizes larger errors more heavily due to the squaring operation.
*   **Characteristics**:
    *   **Sensitivity to Outliers**: Squaring the errors means that larger errors contribute disproportionately more to the total loss, making MSE sensitive to outliers.
    *   **Smooth and Differentiable**: The MSE function is convex, smooth, and differentiable everywhere, which makes it well-suited for gradient-based optimization methods.
    *   **Interpretation**: Represents the average of the squared differences between predicted and actual values. The unit of MSE is the square of the unit of the target variable.

#### More on Mean Absolute Error (MAE):

*   **Formula**: $$\sum_{i=1}^{D}|x_i-y_i|$$ or $$\frac{1}{D}\sum_{i=1}^{D}|x_i-y_i|$$
*   **Use Cases**: Used in **regression problems**, especially when the presence of outliers is a concern. It measures the average magnitude of the errors without considering their direction.
*   **Characteristics**:
    *   **Robustness to Outliers**: Unlike MSE, MAE is less sensitive to outliers because it uses the absolute difference instead of squaring the error. A large error from an outlier doesn't disproportionately affect the total loss.
    *   **Linear Penalty**: It provides a linear penalty for errors, meaning the magnitude of the error directly corresponds to the penalty. This can sometimes make it harder to optimize due to its non-differentiability at zero (though in practice, subgradients are used).
    *   **Interpretation**: Represents the average absolute difference between predicted and actual values. The unit of MAE is the same as the unit of the target variable, making it more interpretable than MSE.

#### More on Hinge Loss:

*   **Formula**: $$max(0, 1 - y \cdot \hat{y})$$ where $y$ is the true label (-1 or 1) and $\hat{y}$ is the predicted score.\
*   **Use Cases**: Primarily used in **binary classification problems**, particularly with Support Vector Machines (SVMs). It aims to maximize the margin between classes.\
*   **Characteristics**: \
    *   **Maximizing Margin**: Hinge loss encourages a correct classification with a certain margin. If a prediction is correct and outside the margin, the loss is zero. If it's within the margin or incorrect, there's a linear penalty.\
    *   **Non-differentiable**: It is not differentiable at y*y_hat = 1, but subgradient methods can still be used for optimization.\
    *   **Suitable for SVMs**: Its design aligns well with the objective of SVMs, which is to find a decision boundary that maximizes the margin between classes.\
    *   **No Probabilistic Interpretation**: Unlike cross-entropy, hinge loss does not directly provide a probabilistic interpretation of the output.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# --- Data Preparation ---
iris = load_iris()
data, labels = iris.data[:, 0:2], iris.data[:, 2].reshape(-1, 1)

# Shuffle
shuffle_order = np.random.permutation(len(labels))
X = data[shuffle_order] / np.amax(data, axis=0)
y = labels[shuffle_order] / np.amax(labels, axis=0)

# --- Cost Function ---
def cost_function(y_pred, y_actual):
    return 0.5 * np.sum((y_actual - y_pred) ** 2)

# --- Neural Network Blueprint ---
class FlexibleNeuralNetwork(object):
    def __init__(self, n0, n1, n2, activation='sigmoid'):
        self.n0 = n0
        self.n1 = n1
        self.n2 = n2
        self.activation = activation.lower()

        # Scale initialization down slightly to stabilize numerical gradients across all functions
        self.W1 = np.random.randn(self.n0, self.n1) * 0.1
        self.W2 = np.random.randn(self.n1, self.n2) * 0.1

    def _activate(self, z):
        if self.activation == 'sigmoid':
            return 1.0 / (1.0 + np.exp(-z))
        elif self.activation == 'tanh':
            return np.tanh(z)
        elif self.activation == 'relu':
            return np.maximum(0, z)
        else:
            raise ValueError(f"Unknown activation function: {self.activation}")

    def predict(self, x):
        # Layer 1 forward computation with selected activation
        self.z1 = np.dot(x, self.W1)
        self.a1 = self._activate(self.z1)

        # Layer 2 forward computation (linear output combination)
        y_pred = np.dot(self.a1, self.W2)
        return y_pred
```

```python
def get_numerical_gradient(net, X, y):
    w_delta = 1e-8
    y_pred_current = net.predict(X)
    error_current = cost_function(y_pred_current, y)

    dw1 = np.zeros_like(net.W1)
    dw2 = np.zeros_like(net.W2)

    # Gradient for Layer 1
    W1_orig = np.copy(net.W1)
    for i in range(net.n0):
        for j in range(net.n1):
            net.W1 = np.copy(W1_orig)
            net.W1[i][j] += w_delta
            dw1[i][j] = (cost_function(net.predict(X), y) - error_current) / w_delta
    net.W1 = W1_orig

    # Gradient for Layer 2
    W2_orig = np.copy(net.W2)
    for i in range(net.n1):
        for j in range(net.n2):
            net.W2 = np.copy(W2_orig)
            net.W2[i][j] += w_delta
            dw2[i][j] = (cost_function(net.predict(X), y) - error_current) / w_delta
    net.W2 = W2_orig

    return dw1, dw2
```

```python
# Training parameters
n_epochs = 1500
learning_rate = 0.01
activation_functions = ['sigmoid', 'tanh', 'relu']
results = {}

# Train models for each activation function
for act in activation_functions:
    # Setting seed here ensures relatively fair comparison comparisons for initialization
    np.random.seed(42)
    net = FlexibleNeuralNetwork(n0=2, n1=3, n2=1, activation=act)
    errors = []

    for epoch in range(n_epochs):
        dw1, dw2 = get_numerical_gradient(net, X, y)

        net.W1 -= learning_rate * dw1
        net.W2 -= learning_rate * dw2

        loss = cost_function(net.predict(X), y)
        errors.append(loss)

    results[act] = errors
    print(f"{act.capitalize()} Final Error: {errors[-1]:.5f}")

# --- Plot the Convergence Curves ---
plt.figure(figsize=(10, 6))
for act in activation_functions:
    plt.plot(results[act], label=f'{act.capitalize()}')

plt.xlabel('Epochs / Iterations')
plt.ylabel('Sum-Squared Error (SSE)')
plt.title('Convergence Comparison of Activation Functions (Numerical Optimization)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#ex-2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# --- Data Preparation ---
iris = load_iris()
data, labels = iris.data[:, 0:2], iris.data[:, 2].reshape(-1, 1)

# Shuffle and normalize data
shuffle_order = np.random.permutation(len(labels))
X = data[shuffle_order] / np.amax(data, axis=0)
y = labels[shuffle_order] / np.amax(labels, axis=0)

# 1. Original Sum Squared Error (SSE) Function
def sum_squared_error(y_pred, y_actual):
    return 0.5 * np.sum((y_actual - y_pred) ** 2)

# 2. New Mean Absolute Error (MAE) Function
def mean_absolute_error(y_pred, y_actual):
    return np.mean(np.abs(y_actual - y_pred))
```

```python
class NeuralNetwork(object):
    def __init__(self, n0, n1, n2):
        self.n0 = n0
        self.n1 = n1
        self.n2 = n2
        # Small weights scaling initialization to keep gradients stable
        self.W1 = np.random.randn(self.n0, self.n1) * 0.1
        self.W2 = np.random.randn(self.n1, self.n2) * 0.1

    def predict(self, x):
        # Layer 1 forward computation (Sigmoid hidden layer)
        z = np.dot(x, self.W1)
        self.a1 = 1.0 / (1.0 + np.exp(-z))

        # Layer 2 forward computation (Linear output layer)
        y_pred = np.dot(self.a1, self.W2)
        return y_pred
```

```python
def get_loss_gradient(net, X, y, loss_type='sse'):
    w_delta = 1e-8

    # Select target evaluation cost metric function
    cost_fn = sum_squared_error if loss_type == 'sse' else mean_absolute_error

    y_pred_current = net.predict(X)
    error_current = cost_fn(y_pred_current, y)

    dw1 = np.zeros_like(net.W1)
    dw2 = np.zeros_like(net.W2)

    # Gradient calculations for Layer 1 weights
    W1_orig = np.copy(net.W1)
    for i in range(net.n0):
        for j in range(net.n1):
            net.W1 = np.copy(W1_orig)
            net.W1[i][j] += w_delta
            dw1[i][j] = (cost_fn(net.predict(X), y) - error_current) / w_delta
    net.W1 = W1_orig

    # Gradient calculations for Layer 2 weights
    W2_orig = np.copy(net.W2)
    for i in range(net.n1):
        for j in range(net.n2):
            net.W2 = np.copy(W2_orig)
            net.W2[i][j] += w_delta
            dw2[i][j] = (cost_fn(net.predict(X), y) - error_current) / w_delta
    net.W2 = W2_orig

    return dw1, dw2
```

```python
n_epochs = 1500
learning_rate = 0.01

# --- Session 1: Train strictly optimizing for Sum Squared Error (SSE) ---
np.random.seed(42)
net_sse = NeuralNetwork(2, 3, 1)
sse_history_sse = []
mae_history_sse = []

for epoch in range(n_epochs):
    dw1, dw2 = get_loss_gradient(net_sse, X, y, loss_type='sse')
    net_sse.W1 -= learning_rate * dw1
    net_sse.W2 -= learning_rate * dw2

    predictions = net_sse.predict(X)
    sse_history_sse.append(sum_squared_error(predictions, y))
    mae_history_sse.append(mean_absolute_error(predictions, y))

# --- Session 2: Train strictly optimizing for Mean Absolute Error (MAE) ---
np.random.seed(42) # Mirror weights initial distributions configurations
net_mae = NeuralNetwork(2, 3, 1)
sse_history_mae = []
mae_history_mae = []

for epoch in range(n_epochs):
    dw1, dw2 = get_loss_gradient(net_mae, X, y, loss_type='mae')
    net_mae.W1 -= learning_rate * dw1
    net_mae.W2 -= learning_rate * dw2

    predictions = net_mae.predict(X)
    sse_history_mae.append(sum_squared_error(predictions, y))
    mae_history_mae.append(mean_absolute_error(predictions, y))
```

```python
plt.figure(figsize=(14, 5))

# Plot 1: SSE Performance Comparison
plt.subplot(1, 2, 1)
plt.plot(sse_history_sse, label='Optimizing for SSE', color='blue')
plt.plot(sse_history_mae, label='Optimizing for MAE', color='orange', linestyle='--')
plt.xlabel('Epochs')
plt.ylabel('Sum Squared Error (SSE)')
plt.title('Performance Measured in SSE')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 2: MAE Performance Comparison
plt.subplot(1, 2, 2)
plt.plot(mae_history_sse, label='Optimizing for SSE', color='blue', linestyle='--')
plt.plot(mae_history_mae, label='Optimizing for MAE', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Mean Absolute Error (MAE)')
plt.title('Performance Measured in MAE')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
#ex-4
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# --- Data Preparation ---
iris = load_iris()
data, labels = iris.data[:, 0:2], iris.data[:, 2].reshape(-1, 1)

# Shuffle and normalize
np.random.seed(42)
shuffle_order = np.random.permutation(len(labels))
X = data[shuffle_order] / np.amax(data, axis=0)
y = labels[shuffle_order] / np.amax(labels, axis=0)

def sum_squared_error(y_pred, y_actual):
    return 0.5 * np.sum((y_actual - y_pred) ** 2)

# --- Extended Neural Network Class ---
class BiasedNeuralNetwork(object):
    def __init__(self, n0, n1, n2):
        self.n0 = n0  # Input nodes
        self.n1 = n1  # Hidden nodes
        self.n2 = n2  # Output nodes

        # Scale initial weights and biases down to 0.01 for stability
        self.W1 = np.random.randn(self.n0, self.n1) * 0.01
        self.W2 = np.random.randn(self.n1, self.n2) * 0.01
        self.b1 = np.random.randn(1, self.n1) * 0.01

    def predict(self, x):
        # NEW: Added self.b1 to the linear combination
        self.z1 = np.dot(x, self.W1) + self.b1
        self.a1 = 1.0 / (1.0 + np.exp(-self.z1))  # Sigmoid activation

        # Output layer (Linear combination remains standard regression output)
        y_pred = np.dot(self.a1, self.W2)
        return y_pred
```

```python
def get_numerical_gradient_with_bias(net, X, y):
    w_delta = 1e-8
    y_pred_current = net.predict(X)
    error_current = sum_squared_error(y_pred_current, y)

    dw1 = np.zeros_like(net.W1)
    dw2 = np.zeros_like(net.W2)
    db1 = np.zeros_like(net.b1)  # Placeholder for bias gradients

    # Gradient for W1
    W1_orig = np.copy(net.W1)
    for i in range(net.n0):
        for j in range(net.n1):
            net.W1 = np.copy(W1_orig)
            net.W1[i][j] += w_delta
            dw1[i][j] = (sum_squared_error(net.predict(X), y) - error_current) / w_delta
    net.W1 = W1_orig

    # Gradient for W2
    W2_orig = np.copy(net.W2)
    for i in range(net.n1):
        for j in range(net.n2):
            net.W2 = np.copy(W2_orig)
            net.W2[i][j] += w_delta
            dw2[i][j] = (sum_squared_error(net.predict(X), y) - error_current) / w_delta
    net.W2 = W2_orig

    # NEW: Gradient for hidden layer bias b1
    b1_orig = np.copy(net.b1)
    for j in range(net.n1):
        net.b1 = np.copy(b1_orig)
        net.b1[0][j] += w_delta
        db1[0][j] = (sum_squared_error(net.predict(X), y) - error_current) / w_delta
    net.b1 = b1_orig

    return dw1, dw2, db1
```

```python
n_epochs = 1500
learning_rate = 0.002

# --- Train Network WITHOUT Hidden Bias ---
np.random.seed(42)
net_no_bias = BiasedNeuralNetwork(2, 3, 1)
net_no_bias.b1 = np.zeros_like(net_no_bias.b1)  # Force bias to 0
errors_no_bias = []

for epoch in range(n_epochs):
    dw1, dw2, _ = get_numerical_gradient_with_bias(net_no_bias, X, y)
    net_no_bias.W1 -= learning_rate * dw1
    net_no_bias.W2 -= learning_rate * dw2

    errors_no_bias.append(sum_squared_error(net_no_bias.predict(X), y))

# --- Train Network WITH Hidden Bias ---
np.random.seed(42)
net_with_bias = BiasedNeuralNetwork(2, 3, 1)
errors_with_bias = []

for epoch in range(n_epochs):
    dw1, dw2, db1 = get_numerical_gradient_with_bias(net_with_bias, X, y)
    net_with_bias.W1 -= learning_rate * dw1
    net_with_bias.W2 -= learning_rate * dw2
    net_with_bias.b1 -= learning_rate * db1  # NEW: Update bias parameters

    errors_with_bias.append(sum_squared_error(net_with_bias.predict(X), y))

# --- Visualizing Results ---
print(f"Final Error (Without Bias): {errors_no_bias[-1]:.5f}")
print(f"Final Error (With Bias):    {errors_with_bias[-1]:.5f}")

plt.figure(figsize=(10, 6))
plt.plot(errors_no_bias, label='Without Hidden Bias', color='crimson', linestyle='--')
plt.plot(errors_with_bias, label='With Hidden Bias', color='teal')
plt.xlabel('Epochs')
plt.ylabel('Sum Squared Error (SSE)')
plt.title('Impact of Hidden Layer Bias on Convergence')
plt.legend()
plt.grid(True, alpha=0.4)
plt.show()
```
